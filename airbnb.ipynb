##Visualize the data

```python
neighbourhood_group = ['Brooklyn','Manhattan','Queens','Manhattan','Brooklyn','Staten Island','Queens','Bronx','Queens','Bronx']
room_type = ['Entire home/apt','Entire home/apt','Private room','Private room','Private room','Entire home/apt','Entire home/apt','Private room','Shared room','Entire home/apt']

room_dict = {}   # create a dictionary named room_dict to store the count of each room type. Loop through the room_type list and increase the count

for i in room_type:
    room_dict[i] = room_dict.get(i , 0) + 1

plt.bar(room_dict.keys(),room_dict.values(),color='green',edgecolor='blue')
plt.title('Room Types',fontsize=15)
plt.xlabel('Room Type',fontsize=15)
plt.ylabel('Count',fontsize=15)
plt.show()
```

We found that Entire home/apt is the highest number of room types overall and prices are high in the **brooklyn** and **Manhattan** for entire home/apt

## **3.What can we learn from predictions? (ex: locations, prices, reviews, etc)**

```python
area_reviews = airbnb.groupby(['neighbourhood_group'])['number_of_reviews'].max().reset_index()
area_reviews
```

```python
area = area_reviews['neighbourhood_group']
review = area_reviews['number_of_reviews']
fig = plt.figure(figsize = (10,5))

plt.bar(area,review,color='Blue',width = 0.5)
plt.xlabel('Area',fontsize=15)
plt.ylabel('Review',fontsize=15)
plt.title('Number of reviews in term of area',fontsize=15)
plt.show()
```

```python
price_area = airbnb.groupby(['price'])['number_of_reviews'].max().reset_index()
price_area.head(10)
```

```python
price_list = price_area['price']
review = price_area['number_of_reviews']
fig = plt.figure(figsize = (10,5))

plt.scatter(price_list,review)
plt.xlabel('Price')
plt.ylabel('Number of reviews')
plt.title('Number of reviews v/s price')
plt.show()
```

From above visulisation we can say that most of people like to stay in less price and their reviews are higher in those areas

## **4.Which hosts are the busiest and why is the reason?**

```python
busy_hosts = airbnb.groupby(['host_id', 'host_name','room_type'])['number_of_reviews'].max().reset_index()
busy_hosts = busy_hosts.sort_values('number_of_reviews',ascending = False).head(10)
busy_hosts
```

```python
name_host = busy_hosts['host_name']
review_got = busy_hosts['number_of_reviews']

fig = plt.figure(figsize = (10,5))

plt.bar(name_host,review_got,color = 'purple' ,width = 0.5)
plt.title("Busiest hosts in term of reviews",fontsize = 15)
plt.xlabel('Name of the host',fontsize = 15)
plt.ylabel('Review',fontsize = 15)
plt.show()
```

## **5. Which Hosts are charging higher price?**

```python
Higher_price = airbnb.groupby(['host_id','host_name','room_type',"neighbourhood_group"])["price"].max().reset_index()
Higher_price = Higher_price.sort_values('price',ascending = False).head(10)
Higher_price
```

```python
name_of_host = Higher_price['host_name']
price_charge = Higher_price['price']
fig = plt.figure(figsize = (10,5))

plt.bar(name_of_host,price_charge,color = 'magenta' , width = 0.5)
plt.xlabel('name of the host',fontsize=15)
plt.ylabel('price',fontsize=15)
plt.title('Hosts with maximum price charge',fontsize=15)
plt.show()
```

##**6. Is there any traffic difference among different areas and what could be the reason for it?**

```python
airbnb.head(10)
```

```python
traffic_areas = airbnb.groupby(['neighbourhood_group','room_type'])['minimum_nights'].count().reset_index()
traffic_areas = traffic_areas.sort_values('minimum_nights',ascending = False).head(10)
traffic_areas
```

```python
area_traffic = traffic_areas['room_type']
room_stayed = traffic_areas['minimum_nights']

fig = plt.figure(figsize = (10,5))

plt.bar(area_traffic,room_stayed,color = 'magenta',width = 0.2)
plt.xlabel('Room Type',fontsize=15)
plt.ylabel('Minimum Nights',fontsize=15)
plt.title('Traffic areas Based on Minimum Nights Booked',fontsize = 15)
plt.show()
```

From the visualisation we found that most of the people likely to stay Entire home and Private room which are present in Manhattan, Brookyln & Queens and also visitors referring stay in room which listing price is less

##**8. What is the room count in overall NYC according to the listing of room types?**

```python
plt.rcParams['figure.figsize'] = (8,5)
ax = sns.countplot(y ='room_type',hue = 'neighbourhood_group',data = airbnb,palette = 'bright')

total = len(airbnb['room_type'])
for p in ax.patches:
        percentage = '{:1f}%'.format(100*p.get_width()/total)
        x = p.get_x() + p.get_width() + 0.02
        y = p.get_y() + p.get_height()/2
        ax.annotate(percentage,(x,y))

plt.title('Count of each room types in NYC')
plt.xlabel('Rooms')
plt.xticks(rotation=90)
plt.ylabel('Room Counts')

plt.show()
```

Manhattan has more listed properties with Entire room / apt around 27% of total listed followed by Brooklyn with around 19.6%

Private rooms are more in Brooklyn as in 20.7% of the total properties followed by Manhattan with 16.3 of them. While 6.9% of private rooms are from Queens.

We can infer that Brookyln, Queens, Bronx has more private room types while Mahattan which has the highest number of listing in entire NYC has more Entire home/apt room types.
